## Method: 
- Separate the text file into different paragraphs of each unit by regular expression. 
- From each paragraph, we will conduct to find unitcode, unit title, synopsis, prerequisites, prohibitions, requirements, outcomes and chief examiners by creating functions as the steps below.

Step 1: Import the 're' module to use regular expressions in python and json module to extract data as a json format

```python
import re
import json
```

Step 2: Create a function that returns a list of paragraphs as well as unitcode and unit title

```python
def find_ucode_utitle(content):
    pattern1 = r'<h1.*?><.*?>(.*?)<.*?> - (.*?)<.*?> - 2019<.*?><.*?>(.*?)(?=<h1.*?><.*?>.*?<.*?> - .*?<.*?> - 2019<.*?><.*?>|\Z)'
    #The pattern1 will match texts which include the content from "<h1 class="banner_med"><span class="unitcode">"
    #by <h1.*?><.*?> to "<div class="hbk-banner-box">" by (?=<h1.*?><.*?>.*?<.*?> - .*?<.*?> - 2019<.*?><.*?>|\Z) 
    #that means matching the content before "<h1 class="banner_med"><span class="unitcode">" 
    #of next paragraph.
    #The pattern1 will also group the the unitcode and untit title by (.*?).
    return re.findall(pattern1, content.read(), re.DOTALL)
    #The function will return the list of elements including unitcode, unit title and the rest of content
    #The element including the rest of content will be used to find synopsis, prerequisites, prohibitions, 
    #requirements, outcomes and chief examiners of each paragraph.
```

Step 3: Create a function that returns a list of synopsis

```python
def find_synopsis(paragraphs):
    
    #Return a list of synopsis but it still includes unecessary information such as "<p>" or 
    #"<span class="unitlink"><a href="/pubs/2019handbooks/units/OCC5060.html">" or
    #"</a></span>" in the content
    all_synopsis = []
    for paragraph in paragraphs:
        pattern2 = r'(?<=<h2 class="hbk-heading">Synopsis</h2>\n<div>\n)(.*?)(?=\n</div>)'
        #The pattern2 will match texts which include the content from after 
        #"<h2 class="hbk-heading">Synopsis</h2>\n<div>" to
        #before "\n</div>" in the content
        all_synopsis.append(re.findall(pattern2, paragraph[2], re.DOTALL))
        #This will return a list of synopsis but still includes unecessary information
    #return all_synopsis
    
    #Remove unecessary information "<p>" but still include unecessary information 
    #like "<span class="unitlink"><a href="/pubs/2019handbooks/units/OCC5060.html">" and "</a></span>"
    separate_synopsis = []
    for synopsis in all_synopsis:
        pattern3 = r'<p>(.*?)</p>'
        #This pattern3 will match any content in each element of all_synopsis except "<p>"
        if synopsis != []:
        #For elements that are not empty in all_synopsis    
            separate_synopsis.append(re.findall(pattern3, synopsis[0], re.DOTALL))
            #This will return a list of synopsis but still includes unecessary information
        else:
        #For elements that are empty, append [] into the list
            separate_synopsis.append([])
    #return separate_synopsis
    
    #Remove unecessary information "<span class="unitlink"><a href="/pubs/2019handbooks/units/OCC5060.html">"
    #but still include unecessary information "</a></span>"
    separate_synopsis1 = []
    for each_synopsis1 in separate_synopsis:
        sample1 = []
        for inner_synopsis1 in each_synopsis1:
            sample1.append(re.sub(r'<span class="unitlink"><a href="/pubs/2019handbooks/units/.*?.html">', '', inner_synopsis1))
            #This pattern above will replace 
            #<span class="unitlink"><a href="/pubs/2019handbooks/units/.*?.html"> by ''
            #Append elements which are any content except 
            #<span class="unitlink"><a href="/pubs/2019handbooks/units/.*?.html"> into an empty list 
        separate_synopsis1.append(sample1)
        #Return a list that includes elements which are any content except
        #<span class="unitlink"><a href="/pubs/2019handbooks/units/.*?.html">
    #return separate_synopsis1
    
    #Remove unecessary information "</a></span>" 
    separate_synopsis2 = []
    for each_synopsis2 in separate_synopsis1:
        sample2 = []
        for inner_synopsis2 in each_synopsis2:
            sample2.append(re.sub(r'</a></span>', '', inner_synopsis2))
            #This pattern above will replace "</a></span>" by ''
            #Append elements which are any content except except "</a></span>"
        separate_synopsis2.append(sample2)
        #Return a list that elements are synopsis of each paragraph
    #return separate_synopsis2

    #Create a list that includes synopsis of each unit. For units that have an empty synopsis, append 'NA'
    separate_synopsis3 = []
    for each_synopsis in separate_synopsis2:
        if each_synopsis == []:
            each_synopsis.append('NA')
            separate_synopsis3.append(each_synopsis)
        else:
            separate_synopsis3.append(each_synopsis)
    return separate_synopsis3
```

Step 4: Create a function that returns prerequisites

```python
def find_prerequisites(paragraphs):
    
    #Return a list including elements are prerequisites but still include unecessary information
    #such as "<p class="hbk-preamble-heading">Prerequisites</p>\n<p>" and "</p>"
    all_prerequisites = []
    for paragraph in paragraphs:
        pattern4 = r'(?<=<p class="hbk-preamble-heading">Prerequisites</p>\n<p>)(.*?)(?=</p>)'
        #The pattern4 will match texts that start from "<p class="hbk-preamble-heading">Prerequisites</p>\n<p>" 
        #to "</p>"
        all_prerequisites.append(re.findall(pattern4, paragraph[2], re.DOTALL))
        #Append these texts above into an empty list
    #return all_prerequisites
    
    #Return a list including elements are prerequisites
    prerequisite_sample = []
    for prerequisite in all_prerequisites:
        if prerequisite != []:
        #Use elements that are not empty in all_prerequisites    
            pattern5 = r'\w{4}\d{3}'
            #The pattern5 will match any prerequisites that have 4 letters first and 3 numbers after
            prerequisite_sample.append(re.findall(pattern5, prerequisite[0], re.DOTALL))
            #Append prerequisites into an empty list
        else:
        #For elements that are empty, appending [] into the list of prerequisites
            prerequisite_sample.append([])
            #Append prerequisites into an empty list
    #return prerequisite_sample
    
    #Return a list including elements are co-requisites but still include unecessary information
    #such as "<p class="hbk-preamble-heading">Prerequisites</p>\n<p>" and "</p>"
    all_co_requisites = []
    for paragraph in paragraphs:
        pattern6 = r'(?<=<p class="hbk-preamble-heading">Co-requisites</p>\n<p>)(.*?)(?=</p>)'
        #The pattern6 will match texts that are from 
        #"(?<=<p class="hbk-preamble-heading">Co-requisites</p>\n<p>)(.*?)(?=</p>)" to "</p>"
        all_co_requisites.append(re.findall(pattern6, paragraph[2], re.DOTALL))
        #Append these texts above into an empty list   
    #return all_co_requisites
    
    #Return a list including elements are co-requisites
    co_requisite_sample = []
    for co_requisite in all_co_requisites:
        if co_requisite != []:
            #Use elements that are not empty in all_co_requisites
            pattern5 = r'\w{4}\d{3}'
            #The pattern5 will match any prerequisites that have 4 letters first and 3 numbers after
            co_requisite_sample.append(re.findall(pattern5, co_requisite[0], re.DOTALL))
            #Append co-requisites into an empty list
        else:
        #For elements that are empty, appending [] into the list of co-requisites
            co_requisite_sample.append([])
     #return co_requisite_sample
    
    #Return a list including elements are prerequisites and co-requisites
    list_list_pre_requisites = []
    i = 0
    while i <= len(prerequisite_sample) - 1:
        list_list_pre_requisites.append(prerequisite_sample[i] + co_requisite_sample[i])
        i += 1
    #return list_list_pre_requisites
    
    #Return a set including elements are prerequisites and co-requisites to remove which elements are the same
    list_set_pre_requisites = []
    for each_pre_requisites1 in list_list_pre_requisites:
        each_pre_requisites1 = set(each_pre_requisites1)
        list_set_pre_requisites.append(each_pre_requisites1)
    pre_requisites = []
    for each_pre_requisites2 in list_set_pre_requisites:
        each_pre_requisites2 = list(each_pre_requisites2)
        pre_requisites.append(each_pre_requisites2)
    #return pre_requisites
    
    #Convert the set above into a list
    new_pre_requisites = []
    for each_pre_requisites in pre_requisites:
        if each_pre_requisites == []:
            each_pre_requisites.append('NA')
        new_pre_requisites.append(each_pre_requisites)
    return new_pre_requisites
```

Step 5: Create a function that returns prohibitions

```python
def find_prohibitions(paragraphs):
    
    all_prohibitions = []
    for paragraph in paragraphs:
        pattern7 = r'<p class="hbk-preamble-heading">Prohibitions</p>\n.*?</p>'
        #This pattern will match texts that are from "<p class="hbk-preamble-heading">Prohibitions</p>"
        #to "</p>"
        all_prohibitions.append(re.findall(pattern7, paragraph[2], re.DOTALL))
        #Append these texts above into an empty list
    #return all_prohibitions
    
    #Return a list including elements are prohibitions (The prohibitions can be the same)
    prohibitions_sample = []
    for prohibition in all_prohibitions:
        if prohibition != []:
        #Use elements that are not empty in all_prohibitions
            pattern8 = r'\w{4}\d{3}'
            #This pattern will match any words which have 4 letters first and 3 numbers after
            prohibitions_sample.append(re.findall(pattern8, prohibition[0], re.DOTALL))
            #Append prohibitions into an empty list
        else:
            prohibitions_sample.append([])
            #For elements that are empty, appending [] into an empty list
    #return prohibitions_sample
    
    #Convert the list of prohibitions above into a set to remove prohibition that are the same
    list_set_prohibitions = []
    for each_prohibition1 in prohibitions_sample:
        each_prohibition1 = set(each_prohibition1)
        list_set_prohibitions.append(each_prohibition1)
    #return list_set_prohibitions
    
    #Convert the set of prohibitions above into a list 
    list_list_prohibitions = []
    for each_prohibition2 in list_set_prohibitions:
        each_prohibition2 = list(each_prohibition2)
        list_list_prohibitions.append(each_prohibition2)
    #return list_list_prohibitions
    
    #For paragraphs that do not have any prohibitions, appending 'NA' into the list
    prohibitions = []
    for each_prohibition3 in list_list_prohibitions:
        if each_prohibition3 == []:
            each_prohibition3.append('NA')
        prohibitions.append(each_prohibition3)
    return prohibitions
```

Step 6: Create a function that returns requirements 

```python
def find_requirements(paragraphs):
    
    #Return a list including elements are requirements of each paragraph but still include unecessary information
    #such as "<p>" and "</p>"
    all_requirements = []
    for paragraph in paragraphs:
        pattern9 = r'(?<=h2 class="hbk-heading">Assessment</h2>\n<div>\n)(.*?)(?=\n</div>)'
        #The pattern9 will match texts that are from after "?<=h2 class="hbk-heading">Assessment</h2>\n<div>\n" 
        #to before "\n</div>" in each paragraph
        all_requirements.append(re.findall(pattern9, paragraph[2], re.DOTALL))
        #Append these content extracted from each paragraph by using the pattern above into a list
    #return all_requirements
    
    #Return a list including elements
    get_requirements = []
    for requirement in all_requirements:
        if requirement != []:
        #For requirements are not empty
            pattern10 = r'(?<=<p>)(.*?)(?=</p>)'
            #The pattern10 will match content of each element in the list all_requirements above from
            #"<p>" to "</p>"
            get_requirements.append(re.findall(pattern10, requirement[0], re.DOTALL))
            #Append these content extracted from the list all_requirements by using the pattern10 above into a list
        else:
            get_requirements.append([])
            #For requirements are empty, appending '[]'
    #return get_requirements

    #For paragraphs that do not have any requirements, appending 'NA' into the list
    requirements = []
    for each_requirement in get_requirements:
        if each_requirement == []:
            each_requirement.append('NA')
        requirements.append(each_requirement)
    return requirements
```

Step 7: Create a function that returns outcomes

```python
def find_outcomes(paragraphs):
    
    #Return a list including elements are outcomes of each paragraph but still include unecessary information
    #such as "<p>", "</p>\n<ol princestart="0" start="1" type="1">\n<li>", "</li>" and "<li>"
    all_outcomes = []
    for paragraph in paragraphs:
        pattern11= r'(?<=<h2 class="hbk-heading">Outcomes</h2>\n<div>)(.*?)(?=\n</ol>)'
        #The pattern11 will match texts that are after "<h2 class="hbk-heading">Outcomes</h2>\n<div>"
        #and before "\n</ol>" in each paragraph
        all_outcomes.append(re.findall(pattern11, paragraph[2], re.DOTALL))
        #Append these content extracted from each paragraph by using the pattern above into a list
    #return all_outcomes
    
    #Return a list including elements are outcomes of each paragraph but still include unecessary information
    #such as <p>" and"</p>\n<ol princestart="0" start="1" type="1">\n<li>"
    get_outcomes = []
    for outcome in all_outcomes:
        if outcome != []:
        #Use elements that are not empty in the list all_outcomes above    
            pattern12 = r'(?<=<li>)(.*?)(?=</li>)'
            #The pattern12 will match texts that are after "<li>" and "</li>" 
            #in each element of the list all_outcomes above
            get_outcomes.append(re.findall(pattern12, outcome[0]))
            #Append these content extracted from each element of the list all_outcomes above into an empty list
        else:
            get_outcomes.append([])
            #For elements are empty, appending '[]' into the list
    #return get_outcomes
    
    #Return a list including elements are outcomes of each paragraph without any unecessary information
    get_outcomes1 = []
    for each_outcome in get_outcomes:
        get_inner_outcome = []
        if each_outcome != []:
            for each_inner_outcome in each_outcome:
                pattern13 = r'\n\n<p>(.*?)</p>'
                #The pattern11 will match texts that are from "\n\n<p>" to "</p>"
                get_inner_outcome.append(re.sub(pattern13, '', each_inner_outcome))
                #The pattern will replace contents that extracted from each element of the list
                #each_outcome by '' and appending them into an empty list
        else:
            get_inner_outcome.append(each_inner_outcome)
        get_outcomes1.append(get_inner_outcome)
    #return get_outcomes1

    #For paragraphs that do not have any outcomes, appending 'NA' into the list
    outcomes = []
    for each_outcome1 in get_outcomes1:
        if each_outcome1 == []:
            each_outcome1.append('NA')
        outcomes.append(each_outcome1)
    return outcomes
```

Step 8: Create a function that returns chief examiners

```python
def find_chief_examiners(paragraphs):
    
    #Return a list including elements are chief examiners but still include unecessary information 
    #such as ""\n<a href="http://staffsearch.monash.edu/?name=Steven Siems">" and "</a>\n<br/>\n</p>\n"
    all_chief_examiners = []
    for paragraph in paragraphs:
        pattern12 = r'(?<=<p class="hbk-highlight-heading">Chief examiner\(s\)</p>\n<p>)(.*?)(?=<p class="hbk-highlight-heading">Coordinator)'
        #The pattern12 will match texts that are from after 
        #"<p class="hbk-highlight-heading">Chief examiner\(s\)</p>\n<p>"
        #to before "<p class="hbk-highlight-heading">Coordinator"
        all_chief_examiners.append(re.findall(pattern12, paragraph[2], re.DOTALL))
        #Append texts that are extracted from each paragraph by the pattern above into an empty list
    #return all_chief_examiners
    
    #Return a list of chief examiners without any unecessary information
    get_chief_examiners = []
    for chief_examiners in all_chief_examiners:
        if chief_examiners != []:
            pattern13 = r'<a.*?>(.*?)</a>'
            #The pattern13 will match texts that are from after "<a.*?>" and "</a>"
            get_chief_examiners.append(re.findall(pattern13, chief_examiners[0], re.DOTALL))
            #Append texts that are extracted from each element of the list all_chief_examiners above 
            #into an empty list
        else:
            get_chief_examiners.append([])
    #return get_chief_examiners
    
    #For paragraphs that do not have any chief examiners, appending 'TBA' into the list
    chief_examiners = []
    for each_chief_examiner in get_chief_examiners:
        if each_chief_examiner == []:
            each_chief_examiner.append('TBA')
        chief_examiners.append(each_chief_examiner)
    return chief_examiners
```

Step 9: Create a function to extract data into an XML format

```python
def xml(file, code_title, synopsis, prerequisites, prohibitions, requirements, outcomes, chief_examiners):
    
    #Write open tag to the xml file
    file.write('<?xml version="1.0" encoding="UTF-8" ?>\n')
    file.write('<units>\n')
    
    for i in range(0, len(code_title)):
        #Write unit id, title, synopsis to the xml file
        file.write(f"<unit id='{code_title[i][0]}'>\n")
        file.write(f'<title> {code_title[i][1]}</title>\n')
        file.write(f'<synopsis> {synopsis[i][0]}</synopsis>\n')

        #Write prerequisites to the xml file
        if prerequisites[i] == ['NA']:
            file.write(f"<pre_requistics> {requirements[i][0]}  </pre_requistics>\n")
        else:
            file.write(f'<pre_requistics>\n')
            for each in prerequisites[i]:
                file.write(f'<pre_requistic>{each}</pre_requistic>')
            file.write(f'</pre_requistics>\n')

        #Write prohibitions to the xml file
        if prohibitions[i] == ['NA']:
            file.write(f"<prohibisions> {prohibitions[i][0]}  </prohibisions>\n")
        else:
            file.write(f'<prohibisions>\n')
            for each in prohibitions[i]:
                file.write(f'<prohibision>{each}</prohibision>')
            file.write(f'</prohibisions>\n')

        #Write requirements to the xml file
        if requirements[i] == ['NA']:
            file.write(f"<requirements> {requirements[i][0]}  </requirements>\n")
        else:
            file.write(f'<requirements>\n')
            for each in requirements[i]:
                file.write(f'<requirement>{each}</requirement>')
            file.write(f'</requirements>\n')

        #Write outcomes to the xml file
        if outcomes[i] == ['NA']:
            file.write(f'<outcomes> {outcomes[i][0]} </outcomes>\n')
        else:
            file.write(f'<outcomes>\n')
            for each in outcomes[i]:
                file.write(f'<outcome>{each}</outcome>')
            file.write(f'</outcomes>\n')

        #Write chief examiners to the xml file
        if chief_examiners[i] == ['TBA']:
            file.write(f'<chief_examiners> {chief_examiners[i][0]} </chief_examiners>\n')
        else:
            file.write(f'<chief_examiners>\n')
            for each in chief_examiners[i]:
                file.write(f'<chief_examiner>{each}</chief_examiner>')
            file.write(f'</chief_examiners>\n')
            
        file.write('</unit>\n')
    
    #Write ending task to the xml file
    file.write('</units>\n')
```

Step 10: Create a function to extract data into an JSON format

```python
#Create a function that creates a dictionary including unit code, unit title, synopsis, prerequisites, prohibitions,
#requirements, outcomes and chief examiners of each unit
def json_file(code_title, synopsis, prerequisites, prohibitions, requirements, outcomes, chief_examiners):
    
    #Create a list of dictionary
    units = []
    for i in range(0, len(code_title)):
        unit = {}
        
        #Appending key and value of id, title to the dictionary
        unit['@id'] = code_title[i][0]
        unit['title'] = code_title[i][1]
        
        #Appending synopsis to the dictionary
        if synopsis[i] == ['NA']:
            unit['synopsis'] = 'NA'
        else:
            unit['synopsis'] = synopsis[i][0]
        
        #Appending prerequisites to the dictionary
        if prerequisites[i] == ['NA']:
            unit['pre_requistics'] = "NA"
        else:
            unit['pre_requistics'] = {
                'pre_requistic': prerequisites[i]
            }
        
        #Appending prohibitions to the dictionary
        if prohibitions[i] == ['NA']:
            unit['prohibitions'] = 'NA'
        else:
            unit['prohibitions'] = {
                'prohibision': prohibitions[i]
            }
            
        #Appending requirements to the dictionary
        if requirements[i] == ['NA']:
            unit['requirements'] = 'NA'
        else:
            unit['requirements'] = {
                'requirement': requirements[i][0]
            }
            
        #Appending outcomes to the dictionary
        if outcomes[i] == ['NA']:
            unit['outcomes'] = 'NA'
        else:
            unit['outcomes'] = {
                'outcome': outcomes[i]
            }
            
        #Appending chief_examiners to the dictionary
        if chief_examiners[i] == ['TBA']:
            unit['chief_examiners'] = 'TBA'
        else:
            unit['chief_examiners'] = {
                'chief_examiner': chief_examiners[i][0]
            }
        
        #Appeding each dictionary into an empty list 
        units.append(unit)
        
    #The dictionary included unit code, unit title, synopsis, prerequisites, prohibitions, 
    #requirements, outcomes and chief examiners of each unit
    units = {
        "units" : {
            "unit": units
        }
    }
    
    return units
```

Step 11: Create a main function including all functions above and the data

```python
def main():
    
    data = open('29911508.txt', 'r')
    #Read information from the text file
    
    separated_paragraph = find_ucode_utitle(data)
    #Return the function that returns a list of paragraphs as well as unitcode and unit title 
    #Put the result of the function find_ucode_utitle(data) into a variable
    print(separated_paragraph)
    
    
    find_synopsis(separated_paragraph)
    #Return the function that returns prohibitions
    
    find_prerequisites(separated_paragraph)
    #Return the function that returns prerequisites
    
    find_prohibitions(separated_paragraph)
    #Return the function that returns prohibitions
    
    find_requirements(separated_paragraph)
    #Return the function that returns requirements
    
    find_outcomes(separated_paragraph)
    #Return the function that returns outcomes
    
    find_chief_examiners(separated_paragraph)
    #Return the function that returns chief examiners
    
    #Create an xml file and extract the content into this file
    with open('29911508.xml', 'w') as f:
        xml(f, separated_paragraph, find_synopsis(separated_paragraph), find_prerequisites(separated_paragraph),
            find_prohibitions(separated_paragraph), find_requirements(separated_paragraph), find_outcomes(separated_paragraph),
            find_chief_examiners(separated_paragraph))
        
    #Create a json file and extract the content into this file
    with open('29911508.json', 'w') as f:
        json.dump(json_file(separated_paragraph, find_synopsis(separated_paragraph), find_prerequisites(separated_paragraph), 
                       find_prohibitions(separated_paragraph), find_requirements(separated_paragraph), find_outcomes(separated_paragraph), 
                       find_chief_examiners(separated_paragraph)), f, indent = 5)

if __name__ == "__main__":
    main()
```
